```r
# Nick Saali 
# ODE Chapter 22
# Sources: R-Bloggers, Quantitative Ecology by Matthiopoulos, and r-project.org

# Van der Pol Equation which has applications in biological and physical 
#    phenomena, specifically phenomena which exhibit oscillations 
#    including many biochemicals that oscilate in their secretion/concentration
#    according to circadian rhythms 

#install.packages('deSolve')

library(deSolve)

# Model implemented as a function here

VanDerPol <- function (t, y, mu) {
list(c(    y[2],
mu * (1 - y[1]^2) * y[2] - y[1]))}

# Initial Conditions of the State Variables 
y_initial <- c(y1 = 2, y2 = 0)

# "Stiffness" indicates regions where the solution changes very slowly
# and appears "sharp" and "stiff"
# parms = mu 

stiff <- ode(y = y_initial, func = VanDerPol,
times = 0:3000, parms = 1000)

nonstiff <- ode(y = y_initial, func = VanDerPol,
times = seq(0, 30, by = 0.01),
parms = 1)

# Return Matrix 
head(stiff, n = 5)
plot(stiff, type = "l", which = "y1",
lwd = 2, ylab = "y",
main = "Initial Value ODE, mu = 1000 (stiff) ")
plot(nonstiff, type = "l", which = "y1",
lwd = 2, ylab = "y",
main = "Initial Value ODE, mu = 1 (non-stiff) ") 
```
