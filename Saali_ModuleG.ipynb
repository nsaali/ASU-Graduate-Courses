```r
#Nick Saali - Module G - Numerical Integration with R

# Trapezoid Rule

## install.packages('ggplot2')
library(ggplot2)

# ∫ a=0 b=π/4 xsinx dx    in R  ...


f <- function(x) { return(x * sin(x))}

# store data for approximation 
new.data <- data.frame(cbind(c(0, pi/4, pi/4, 0), c(0, f(pi/4), 0, 0)))

# plot fxn and approximation

ggplot(data = data.frame(x = 0), mapping = aes(x = x)) + 
  stat_function(fun = f, size = 1.05, alpha = 0.75, color='green') +  geom_segment(aes(x = 0, y = 0, xend = pi/4, yend = f(pi/4))) + 
  geom_segment(aes(x = pi/4, y = 0, xend = pi/4, yend = f(pi/4))) +  geom_polygon(data = new.data, aes(x=X1, y=X2), fill = 'green', alpha = 0.2) + 
  geom_area(stat = 'function', fun = f, fill = 'black', alpha = 0.3, xlim = c(0, pi/4)) + 
  xlim(c(-0.5,1))


```
