```r
# Nick Saali - Module H - ODE

# Lotka - Volterra Equations describing Predator-Prey Dynamics

#   install.packages('deSolve')
library(deSolve)

LotVol <- function (Time, State, Parameters) {
    with(as.list(c(State, Parameters)), {
        dx = x*(alpha - beta*y)
        dy = -y*(gamma - delta*x)
        
        return(list(c(dx, dy))) }) }

# alpha = prey growth rate (linear)
# beta = prey eaten by predator rate
# delta = predator 
# gamma = predator death rate

# Assumes predator death rate higher (population decay rate) than prey population growth

Parameters <- c(alpha = 1.5, beta = .5, gamma = .2, delta = .6)
State <- c(x = 10, y = 10)
Time <- seq(0, 100, by = 1)

out <- as.data.frame(ode(func = LotVol, y = State, parms = Parameters, times = Time))

matplot(out[,-1], type = "l", col = 1, xlab = "time", ylab = "population")

legend("topright", c("Predator", "Prey"), lty = c(1,2))
```
