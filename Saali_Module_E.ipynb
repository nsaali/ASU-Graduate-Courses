```python
# Nick Saali 
#  Module D - Taylor Series and Approximation


from sympy.abc import x
from sympy import series
from sympy.utilities.lambdify import lambdify

import numpy as np
import matplotlib.pyplot as plt
import math

# Example : f(x) = e^x

# First, note that f(x) = f'(x) = ... = f^n(x) = e^x
# If we were to use x = 0 as a reference point, Taylor Expansions are known
#   instead as Maclaurin Expansions.

my_fxn = math.e**x 
TaylorAppx8 = series(my_fxn, n=8).removeO(); TaylorAppx7 = series(my_fxn, n=7).removeO()
TaylorAppx6 = series(my_fxn, n=6).removeO(); TaylorAppx5 = series(my_fxn, n=5).removeO()
TaylorAppx4 = series(my_fxn, n=4).removeO(); TaylorAppx3 = series(my_fxn, n=3).removeO()
TaylorAppx2 = series(my_fxn, n=2).removeO(); TaylorAppx1 = series(my_fxn, n=1).removeO()

Eval_my_fxn = lambdify(x, my_fxn, modules=['numpy'])
Eval_TaylorAppx8 = lambdify(x, TaylorAppx8, modules=['numpy']); Eval_TaylorAppx7 = lambdify(x, TaylorAppx7, modules=['numpy'])
Eval_TaylorAppx6 = lambdify(x, TaylorAppx6, modules=['numpy']); Eval_TaylorAppx5 = lambdify(x, TaylorAppx5, modules=['numpy'])
Eval_TaylorAppx4 = lambdify(x, TaylorAppx4, modules=['numpy']); Eval_TaylorAppx3 = lambdify(x, TaylorAppx3, modules=['numpy'])
Eval_TaylorAppx2 = lambdify(x, TaylorAppx2, modules=['numpy']); Eval_TaylorAppx1 = lambdify(x, TaylorAppx1, modules=['numpy'])

a = np.linspace(-4, 2)
plt.plot(a, Eval_my_fxn(a), 'k', label='f(x) = e^x')
plt.plot(a, Eval_TaylorAppx8(a), 'navy', label='n=8')
plt.plot(a, Eval_TaylorAppx7(a), 'darkblue', label='n=7')
plt.plot(a, Eval_TaylorAppx6(a), 'mediumblue', label='n=6')
plt.plot(a, Eval_TaylorAppx5(a), 'blue', label='n=5')
plt.plot(a, Eval_TaylorAppx4(a), 'darkviolet', label='n=4')
plt.plot(a, Eval_TaylorAppx3(a), 'purple', label='n=3')
plt.plot(a, Eval_TaylorAppx2(a), 'm', label='n=2')

plt.title('Taylor Approximation Increasing Accuracy for f(x) = e^x')
plt.legend(loc='best')
plt.show()

b = np.linspace(0, 5)
plt.plot(b, Eval_my_fxn(b), 'k', label='f(x) = e^x')
plt.plot(b, Eval_TaylorAppx8(b), 'navy', label='n=8')
plt.plot(b, Eval_TaylorAppx7(b), 'darkblue', label='n=7')
plt.plot(b, Eval_TaylorAppx6(b), 'mediumblue', label='n=6')
plt.plot(b, Eval_TaylorAppx5(b), 'blue', label='n=5')
plt.plot(b, Eval_TaylorAppx4(b), 'darkviolet', label='n=4')
plt.plot(b, Eval_TaylorAppx3(b), 'purple', label='n=3')
plt.plot(b, Eval_TaylorAppx2(b), 'm', label='n=2')

plt.title('Taylor Approximation Increasing Accuracy for f(x) = e^x')
plt.legend(loc='best')
plt.show()


```
